```python
import tensorflow as tf
from matplotlib import pyplot as plt
import numpy as np
image_name = "VOCdevkit/VOC2012/JPEGImages/2007_000032.jpg"
img_name_tensor = tf.convert_to_tensor(image_name,dtype=tf.string)

img_contents = tf.read_file(img_name_tensor)

image = tf.image.decode_jpeg(img_contents, channels=3)
sess = tf.Session()
image_array = sess.run(image)
# plt.imshow(result[0, :, :, 0])
# plt.show()
```

```python
IMG_MEAN = np.array((116.66876762,104.00698793,122.67891434), dtype=np.float32)
img = image_array-IMG_MEAN
```

```python
plt.imshow(image_array)
plt.show()
```

```python
img_rez = tf.image.resize_image_with_crop_or_pad(img, 320, 320)
```

```python
resImg = sess.run(img_rez)
```
